```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
missing = ['N/A', 'NA', np.nan,'na']
df = pd.read_csv('country_vaccinations.csv', na_values = missing)
```

```python
df.head(15)
```

<b><font size = "6">Vaccines used in various countries

```python
ls=df.country
ls1=df.vaccines
ls2=[]
for x in ls1:
    ls2.append(x.split(','))
dict={ls[0]:ls2[0]}
for x in range(1,len(ls)):
    if not ls[x] in dict:
        dict[ls[x]]=ls2[x]
dict
```

<b><font size = "6">Case Study 1: India

```python
filt = (df['country'] == 'India')
a = df[filt]
ind = pd.DataFrame(a)
ind.daily_vaccinations.fillna(0, inplace = True)
ind.rename(columns={'date': 'date(2021)'}, inplace = True)
ind['date(2021)']=ind['date(2021)'].str[5:]
ind
```

```python
y = ind['daily_vaccinations']
x = ind['date(2021)']
plt.style.use('seaborn-whitegrid')
plt.plot(x, y)
plt.xticks(rotation = 90)
plt.autoscale(tight = 0)
plt.xlabel('Dates in 2021')
plt.ylabel('Daily Vaccinations')
plt.title('Daily vaccinations')
plt.tight_layout()
```

<b>Total number of vaccinations

```python
sum1 = 0
for x in ind['daily_vaccinations']:
    sum1 = sum1 + x
print(sum1)
```

<b>Vaccines use in India

```python
print(ind.vaccines.unique())
```
